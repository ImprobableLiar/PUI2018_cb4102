```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import math
%matplotlib inline
```

```python
np.random.seed(42) #set the seed for reproducibility
```

```python
sample_size = 100 #set the number of random samples for each distribution
my_mean = 100
```

```python
#initialize list to hold samples for each distribution

gaussian_samples = [] 
poisson_samples = []
binomial_samples = []
chi_squared_samples = []
uniform_samples = []
```

```python
def sample_gen():
    """function to create a random sample size between 10 and 2000"""
    return np.random.randint(10, 2000)
```

```python

#generate samples for each distribution

for i in range(sample_size):
    sample = np.random.randn(sample_gen())+my_mean
    gaussian_samples.append(sample)
    
for i in range(sample_size):
    sample = np.random.poisson(lam=my_mean,size=sample_gen()) 
    poisson_samples.append(sample)  

for i in range(sample_size):
    sample = np.random.binomial(my_mean/0.5,0.5,sample_gen()) 
    binomial_samples.append(sample) 
    
for i in range(sample_size):
    sample = np.random.chisquare(my_mean, sample_gen())
    chi_squared_samples.append(sample) 
    
for i in range(sample_size):
    sample = np.random.uniform(my_mean/2,my_mean*1.5, sample_gen())
    uniform_samples.append(sample) 
```

```python
#check the length for each distribution to make sure there are 100 samples in each
print(len(gaussian_samples), len(poisson_samples), len(binomial_samples), len(chi_squared_samples), len(uniform_samples))
```

```python
gaussian_samples[0].mean()
```

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
for i in range(sample_size):
    ax.scatter(gaussian_samples[i].size, gaussian_samples[i].mean())
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Mean of Samples from Gaussian Distribution")
```

## Figure 1: Mean of Samples from a Gaussian Distribution
### The figure illustrates the Central Limit Theorum and the law of large numbers because as the sample size increases we see a reversion to the expected mean of 100. In contrast, smaller sample sizes have larger dispersion from the mean.

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
for i in range(sample_size):
    ax.scatter(poisson_samples[i].size, poisson_samples[i].mean())
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Mean of Samples from Poisson Distribution")
```

## Figure 2: Mean of Samples from a Poisson Distribution

### The means from the poisson distribution reinforces the law of large numbers - as the sample size increases mean values begin to converge to the expected mean. However we do see a larger dispersion for the Poisson distribution at values above 1000 than we do for the other distributions.

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
for i in range(sample_size):
    ax.scatter(binomial_samples[i].size, binomial_samples[i].mean())
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Mean of Samples from Binomial Distribution")
```

## Figure 3: Mean of Samples from a Binomial Distribution

### The sample means from the binomial distribution exhibit the same behavior of the poisson and gaussian distributions in that as the size of the sample grows we see a convergence of the sample mean to the value expected (100)

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
for i in range(sample_size):
    ax.scatter(chi_squared_samples[i].size, chi_squared_samples[i].mean())
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Mean of Samples from Chi-Squared Distribution")
```

## Figure 4: Mean of Samples from a Chi-Squared Distribution

### The sample means from the Chi-square distribution shows high levels of dispersion in sample sizes below 100 (and especially below 50) but then converges towards the expected mean about 1000. As expected convergence to the expected mean increases as sample size increases.

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
for i in range(sample_size):
    ax.scatter(uniform_samples[i].size, uniform_samples[i].mean())
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Mean of Samples from Uniform Distribution")
```

## Figure 5: Mean of Samples from a Uniform Distribution

### As with the other distirubtions we see high levels of dispersion from the mean below 100 and as the sample size grows towards 2000 (maximum allowable in this sample) we see a convergence to the expected mean of 100, reinforcing the law of large numbers and the Central Limit Theorum

```python
#create lists of the mean's from the distributions for each distribution.

gaussian_mean_list = [] #initialize a list for the sample means of the gaussian distribution

for i in range(sample_size):
    sample_mean = gaussian_samples[i].mean() #loops through array of gaussian samples and stores mean in a new list
    gaussian_mean_list.append(sample_mean)

poisson_mean_list = []

for i in range(sample_size):
    sample_mean = poisson_samples[i].mean()
    poisson_mean_list.append(sample_mean)

binomial_mean_list = []

for i in range(sample_size):
    sample_mean = binomial_samples[i].mean() 
    binomial_mean_list.append(sample_mean)

chi_squared_mean_list = []

for i in range(sample_size):
    sample_mean = chi_squared_samples[i].mean() 
    chi_squared_mean_list.append(sample_mean)

uniform_mean_list = []

for i in range(sample_size):
    sample_mean = uniform_samples[i].mean() 
    uniform_mean_list.append(sample_mean)
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(gaussian_mean_list, bins=20, alpha=0.7, label="gaussian") #add gaussian list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Gaussian Distribution")
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(poisson_mean_list, bins=20, alpha=0.6, label="poisson", density=True) #add Poison list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Poisson Distribution")
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(binomial_mean_list, bins=20, alpha=0.4, label="binomial") #add binomial list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Binomial Distribution")
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(chi_squared_mean_list, bins=20, alpha=0.3, label="chi-squared") #add chi-sqauared list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Chi-Squared Distribution")
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(uniform_mean_list, bins=20, alpha=0.2, label="uniform") #add uniform list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Uniform Distribution")
```

```python
aggregate_samples = np.concatenate((uniform_mean_list,chi_squared_mean_list,binomial_mean_list,gaussian_mean_list,poisson_mean_list),axis=0) #combine all distributions so
```

```python
fig = plt.figure(figsize=(10,10)) #initialize figure
ax = fig.add_subplot(111) #initialize subplot

ax.hist(aggregate_samples, bins=40, alpha=0.2, label="uniform") #add uniform list

plt.xlabel("Mean of Sample") #x label added
plt.ylabel("N") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("All Distributions")
```

## Figure 6:

#### I plotted the means of all five distributions separately and all of the distributions appear to converge to the normal distribution, as we would expect from the Central Limit Theorum. When we aggregate the mean's from all the distribution in the final histogram we see even greater convergence to the mean, which is intuitive since sample size has increased (law of large numbers)
