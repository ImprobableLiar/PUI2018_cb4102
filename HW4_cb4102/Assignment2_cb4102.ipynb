### IDEA: Men are more likely to start trips between the hours of 8:00PM - 4:00AM than women

### null hypothesis: The proportion of men who start trips between the hours of 8:00PM - 4:00AM  is greater than the proportion of men who start trips during all other hours of the day

### alternative hypothesis: The proportion of women who start trips between 8:00PM - 4:00AM  is lower than the proportion of women who start trips during all other hours of the day

### Mathematically:

#### H0: (# of men starting trips between 8PM-4AM) / (total # of people starting trips between 8PM-4AM) > (# of men starting trips between 4AM-7:59PM) / (total # of people starting trips between 4AM-7:59PM)

#### H1:  (# of women starting trips between 8PM-4AM) / (total # of people starting trips between 8PM-4AM) <= (# of women starting trips between 4AM-7:59PM) / (total # of people starting trips between 4AM-7:59PM)


### confidence level = 0.05

```python
citibike = pd.read_csv(os.getenv("PUIDATA") + "/" + "201711-citibike-tripdata.csv")
```

```python
citibike.head()
```

```python
citibike['date'] = pd.to_datetime(citibike["starttime"])
```

```python
citibike['hour'] = pd.DatetimeIndex(citibike['starttime']).hour
```

```python
citibike.head()
```

```python
citibike['hour'].groupby(citibike['hour']).count() #check to see the distribution of rides by hour
```

```python
def night_day_label(row):
    """Function to set up a column based on conditions"""
    if row['hour'] < 5:
        return 1
    if row['hour'] >= 20:
        return 1
    else:
        return 0
```

```python
citibike['night_flag'] = citibike.apply(lambda row: night_day_label(row),axis=1) #apply night_day_label
```

```python
citibike[['hour','night_flag']].groupby(['hour','night_flag']).count() #check to make sure the hours are flagged properly
```

```python
citibike_clean = citibike[['hour','gender','night_flag']] #filter and create new dataframe
citibike_clean.head() #print new df head
```

```python
#plot data I think we be relevant:

men_biking = citibike_clean[citibike['gender'] == 1]
women_biking = citibike_clean[citibike['gender'] == 2]
```

```python

fig = plt.figure(figsize=(10,10)) #initialize figure
ax1 = fig.add_subplot(111) #initialize subplot

ax1.hist(men_biking['hour'], bins=20, alpha=0.7, label="male")
ax1.hist(women_biking['hour'], bins=20, alpha=0.5, label="female")

plt.xlabel("Hour of Day") #x label added
plt.ylabel("Number of Rides") #add Y label
plt.legend(loc = "upper right") #add legend
plt.title("Number of Citibike rides by men and women by hour of day")
```
